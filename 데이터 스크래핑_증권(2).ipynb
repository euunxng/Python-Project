```python
import requests
from bs4 import BeautifulSoup
import time
import random
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import numpy as np
```

```python
url = requests.get('https://finance.naver.com/sise/dividend_list.naver?&page=1')
html = BeautifulSoup(url.text)
```

```python
table = html.find('table', class_ = 'type_1 tb_ty')
table = pd.read_html(str(table), header = 1)[0] #1번째를 열이름으로 지정하겠다 ! // 바로 데이터 프레임으로 만들 수 있음~~
table = table.dropna(subset='종목명')
table
```

```python
table.info() #배당금,1년전,2년전,3년전이 실수인지 확인 (-가 껴잇어서 object로 인식된거임!)
```

```python
table['3년전'] = table['3년전'].str.replace('-', '0').astype('int32')
```

```python
table['2년전'] = table['2년전'].str.replace('-', '0').astype('int32')
```

```python
table.info()#모든페이지를 데이터프레임으 가져와서 info로 확인하고 필요할 경우에 하도록 하자.
```

```python
end = int(html.find('td', class_ = 'pgRR').find('a')['href'].split('=')[-1]) #text사용 x
end
```

```python
from tqdm import tqdm

box = []
for i in tqdm(range(1, end+1)):
    url = requests.get(f'https://finance.naver.com/sise/dividend_list.naver?&page={i}')
    html = BeautifulSoup(url.text)

    table = html.find('table', class_ = 'type_1 tb_ty')
    table = pd.read_html(str(table), header = 1)[0]
    table = table.dropna(subset='종목명')
    box.append(table)
    time.sleep(1+random.random())
```

```python
allo = pd.concat(box, ignore_index=True)
allo.info()
```

```python
allo['2년전'] = allo['2년전'].replace('-', 0).astype('int32')
allo['3년전'] = allo['3년전'].replace('-', 0).astype('int32')
```

```python
allo.info()
```

```python
allo2 = allo[(allo['3년전']<allo['2년전']) & (allo['2년전']<allo['1년전']) & (allo['1년전']<allo['배당금'])]
```

```python
allo2
```

```python
from datetime import datetime
today = str(datetime.today().date())

text = f'{today}기준 3년 연속 배당금이 상승한 기업 리스트\n\n'

for i in range(len(allo2)):
    row = allo2.iloc[i]
    name=row['종목명']
    a,b,c,d = row['배당금'],row['1년전'],row['2년전'],row['3년전']
    text += f'{name} : {int(d):,}원 -> {int(c):,}원 -> {int(b):,}원 -> {int(a):,}원\n'

f=open('3년 연속 배당금 상승 기업.txt', 'wt')
f.write(text)
f.close()
```
