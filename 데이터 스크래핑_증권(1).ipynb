```python
import requests
from bs4 import BeautifulSoup
import time
import random
```

```python
url = requests.get('https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page=1')
html = BeautifulSoup(url.text)
```

```python
# 표 - table로 되어 있음!
table = html.find('table', class_ = 'type_2')
table
```

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

table = pd.read_html(str(table))[0] #데이터프레임이 리스트안에 들어이는 형태 [표]-> [0]하면 데이터프레임으로 꺼낼 수 있음
del table['N']
del table['토론실']
```

```python
table = table.dropna(subset='종목명')
table
```

```python
def make_table(html):
    table = html.find('table', class_ = 'type_2')
    table = pd.read_html(str(table))[0]
    del table['N']
    del table['토론실']
    table = table.dropna(subset='종목명')
    return table
```

```python
kospi_page = int(html.find('td', class_ = 'pgRR').find('a')['href'].split('=')[-1]) #text사용 x
kospi_page
```

```python
#퀴즈 : 모든 코스피 페이지를 가져와서 하나의 데이터프레임 만들기!
from tqdm import tqdm
box = []
for i in tqdm(range(1, kospi_page+1)):
    url = requests.get(f'https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page={i}')
    html = BeautifulSoup(url.text)
    table = make_table(html)
    box.append(table)
    time.sleep(1+random.random())
```

```python
kospi = pd.concat(box, ignore_index=True)
kospi['소속'] = 'KOSPI'
kospi.to_excel('kospi.xlsx', index=False)
kospi
```

```python
# 퀴즈: 코스닥까지 모두 가져와서 코스피+코스닥을 합친 하나의 데이터프레임 만들기
# pbr<1 -> 주식 본 가격보다 더 낮다고 판단(저평가된 주식)
# 0<PER<10 행만 추출해서 저평가주식을 찾는다~
```

```python
url = requests.get('https://finance.naver.com/sise/sise_market_sum.naver?sosok=1&page=1')
html = BeautifulSoup(url.text)
```

```python
kosdaq_page = int(html.find('td', class_ = 'pgRR').find('a')['href'].split('=')[-1]) #text사용 x
kosdaq_page
```

```python
def make_table(html):
    table = html.find('table', class_ = 'type_2')
    table = pd.read_html(str(table))[0]
    del table['N']
    del table['토론실']
    table = table.dropna(subset='종목명')
    return table
```

```python
box2 = []
for i in tqdm(range(1, kosdaq_page+1)):
    url = requests.get(f'https://finance.naver.com/sise/sise_market_sum.naver?sosok=1&page={i}')
    html = BeautifulSoup(url.text)
    table = make_table(html)
    box2.append(table)
    time.sleep(1+random.random())
```

```python
kosdaq = pd.concat(box2, ignore_index=True)
kosdaq['소속'] = 'KOSDAQ'
kosdaq
```

```python
stock = pd.concat([kospi, kosdaq], ignore_index=True)
stock.to_excel('total.xlsx', index=False)
```
